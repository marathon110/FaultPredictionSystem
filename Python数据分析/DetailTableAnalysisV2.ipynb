```python
def date2num(dates):
    indexs = []
    for d in dates:
        month = int(d.split('-')[1])
        day = int(d.split('-')[2])
        if month == 9:
            num = day
        elif month == 10:
            num = day+30
        elif month == 11:
            num = day+61
        indexs.append(num-15)
    return indexs
```

```python
nums = date2num(index)
print(nums)
```

```python
device_2400_day['dateNum'] = nums
device_2400_day.head()
```

```python
label = []
for d in device_2400_day['dateNum']:
    if d > 50: label.append(1)
    else: label.append(0)
device_2400_day['lable'] = label
device_2400_day.head()
```
